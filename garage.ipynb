```python
from IPython.display import Image
from time import sleep
from IPython.display import clear_output as clear

enter_image = 'parking_1.png'
exit_image = 'parking_2.png'

class Garage():
    def __init__(self):

        self.spots = {'sedan' : 100,
                     'truck' : 50,
                     'bike' : 20,
                     'EV' : 20,}
        self.rate = {'sedan' : 10,
                     'truck' : 15,
                     'bike' : 8,
                     'EV' : 20,}
        self.tickets = {'sedan' : 0,
                     'truck' : 0,
                     'bike' : 0,
                     'EV' : 0,}
        
        self.type_ = ''
    
    def enter(self):
        vehicle = input ("\nWhat type of vehicle will you be parking today? ['sedan']['truck'/'SUV']['bike']['EV']\n").lower()
        self.type_ = vehicle
        if vehicle == 'sedan':
            print(f"Available spots: {self.spots['sedan']}\n")
            if self.spots['sedan'] > 0:
                print(f"The hourly rate for sedans is ${self.rate['sedan']}\n")
                self.spots['sedan'] -= 1
                self.tickets['sedan'] += 1
                print(f"Your ticket number is {self.tickets['sedan']}\n")
                print("Please take your ticket!")
            else:
                print("Sorry, we can't fit any more sedans right now!")

        elif vehicle == 'truck' or vehicle == 'suv':
            print(f"Available spots: {self.spots['truck']}\n")
            if self.spots['truck'] > 0:
                print(f"The hourly rate for trucks / SUVs is {self.rate['truck']}\n")
                self.spots['truck'] -= 1
                self.tickets['truck'] += 1
                print(f"Your ticket number is {self.tickets['truck']}\n")
                print("Please take your ticket!")
                                     
            else:
                print("Sorry, we can't fit any more trucks / SUVs right now!")

        elif vehicle == 'bike': 
            print(f"Available spots: {self.spots['bike']}")
            if self.spots['bike'] > 0:
                print(f"The hourly rate for bikes is {self.rate['bike']}")
                self.spots['bike'] -= 1
                self.tickets['bike'] += 1
                print(f"Your ticket number is {self.tickets['bike']} ")
                print("Please take your ticket!")
                                            
            else:
                print("Sorry, we can't fit any more bikes right now!")

        elif vehicle == 'ev': 
            print(f"Available spots: {self.spots['EV']}\n")
            if self.spots['EV'] > 0:
                print(f"The hourly rate for electric vehicles is {self.rate['EV']}\n")
                self.spots['EV'] -= 1
                self.tickets['EV'] += 1
                print(f"Your ticket number is {self.tickets['EV']}\n")
                print("Please take your ticket!")
            else:
                print("Sorry, we can't fit any more electric vehicles right now!")
        else: 
            print("Not a valid input. Please pick from the list of vehicle types!")

        sleep(6)
        clear()

        
        
    def pay(self):
        self.hours = float(input("\nHow many hours did you park for? Now, don't lie, you cheapo!!\n"))
        if self.hours > 8:
            self.parking_cost = 100
        else:
            self.parking_cost = float(self.hours) * int(self.rate[self.type_])  
        print(f"Your parking cost is ${self.parking_cost}. Please pay before you leave! ")
        sleep(5)
        print(f"Thank you for your payment of {self.parking_cost}!")
        sleep(2)
        clear()

    def exit(self):
        if exit_image:
            display(Image(exit_image))
        else:
            print("\nThanks for hanging with us today! Let's do it again sometime!")
#         sleep(5)
#         clear()
        
    def lost_ticket(self):
        print("\nSOL... Have fun walking home!!\n") 
        sleep(4)
        print("Just kidding! Please pay max daily rate of $100!\n")   
        sleep(5)
        print("Thank you for your payment of $100!")
        sleep(2)
        clear()
  


    def run(self):
        if enter_image:
            display(Image(enter_image))
        else:
            print("Welcome East & West Parking! Home of all your parking needs!")
        my_car.enter()
        while True:
            action = input("\nHow can I help you? \nOptions:\nPay\nLost Ticket\nHelp\nExit\n").lower()
            if action == 'pay':
                my_car.pay()
            elif action == 'lost ticket':
                my_car.lost_ticket()
            elif action == 'help':
                print("\nFor damages on your car or loss of property, please call 911.\n\nFor any other issues, please contact Garage Management at 800-555-5555.\n")
                sleep(5)
                clear()
            elif action == 'exit':
                response = input("Did you pay your bill? BE HONEST!!! y/n \n")
                if response == 'n':
                    print("Please pay now!") 
                    my_car.pay()
                    sleep(4)
                    my_car.exit()
                elif response == 'y':
                    my_car.exit()
                    sleep(4)
                break
            else:
                print("Invalid option. Please choose one of the options above! ")
                
```
